```python
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
with_mask = with_mask.reshape(200, 50 * 50 * 3)
without_mask = without_mask.reshape(200, 50 * 50 * 3)
```

```python
without_mask.shape
```

```python
with_mask.shape
```

```python
x = np.r_[with_mask, without_mask]
```

```python
x.shape
```

```python
labels = np.zeros(x.shape[0])
```

```python
labels[200: ] = 1.0
names = {0: 'Mask', 1: 'No Mask'}
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, labels, test_size = 0.20) 
```

```python
x_train.shape
```

```python
pca = PCA(n_components=3)
x_train = pca.fit_transform(x_train)
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
svm = SVC()
svm.fit(x_train, y_train)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
haar_data = cv2.CascadeClassifier('data.xml')
capture = cv2.VideoCapture(0)
data = []
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag, img = capture.read()
    if flag:
        face = haar_data.detectMultiScale(img)
        for x,y,w,h in face:
            user_face = img[y:y+h, x:x+w, :]
            user_face = cv2.resize(user_face, (50,50))
            user_face= user_face.reshape(1, -1)
            user_face = pca.transform(user_face)
            pred = svm.predict(user_face)
            n = names[int(pred)]
            if n == 'Mask':
                cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
                cv2.putText(img, n, (40,50), font, 1, (0,255,0), 2)
            else:
                cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,255), 2)
                cv2.putText(img, n, (40,50), font, 1, (0,0,255), 2)
            print(n)
        cv2.imshow("result", img)
        if cv2.waitKey(2) == 27:
            break
capture.release()
cv2.destroyAllWindows()
```
